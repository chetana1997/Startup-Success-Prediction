```python
from datetime import timedelta
days_yr = timedelta(days=365.25)
days_yr
```

```python
%%time
# Creating the feature age for the companies
age = []

for i in range(df_object.shape[0]):
    if (df_object.status.iloc[i] == 'operating') or (df_object.status.iloc[i] == 'acquired') or (df_object.status.iloc[i] == 'ipo'):
        age.append((BASE_YEAR - df_object.founded_at.iloc[i])/days_yr)
    else:
        age.append((df_object.closed_at.iloc[i] - df_object.founded_at.iloc[i])/days_yr)
```

```python
df_object.loc[:, 'age'] = age
```
